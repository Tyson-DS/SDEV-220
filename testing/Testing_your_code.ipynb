```python
def test_sum():
    assert sum([1,2,3]) == 6, "should be six"

def test_sum_tuple():
    assert sum((1,2,2)) == 6, "should be six"

if __name__ == "__main__":
    test_sum()
    test_sum_tuple()
    print("everything passed!")
```

```python
import unittest

class TestSum(unittest.TestCase):
    def test_sum(self):
        self.assertEqual(sum([1,2,3]), 6, "should be six")

    def test_sum_tuple(self):
        self.assertEqual(sum((1,2,3)),6, "should be six")

if __name__ == "__main_":
    unittest.main()
```
