```python
class Solution:
    def sort012(self, arr):
        new_list = []
        for val in (0,1,2):
            for num in arr:
                if num == val:
                    new_list.append(num)
        
        for i in range(len(arr)):
            arr[i] = new_list[i]
        return arr

        
sol = Solution()
arr = [0, 1, 2, 0, 1, 2]
print(sol.sort012(arr))
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1

        while low <= high:
            mid = (low + high) //2

            if arr[mid] == k:
                return mid
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1
        return -1
sol = Solution()
arr = [1,2,3,4,5]
k = 4
sol.binarysearch(arr, k)
```
